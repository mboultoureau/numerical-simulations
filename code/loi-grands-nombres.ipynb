## Partie 1
### 1.1. Génération de nombres aléatoires pour une loi discrète et une loi continue

```python
import numpy as np
import matplotlib.pyplot as plt

# Génération de nombres aléatoires suivant une loi normale
def loi_normale(n):
    # On génère n nombres aléatoires suivant une loi normale
    # de moyenne 0 et d'écart-type 1
    x = np.random.normal(0, 1, n)

    return x

# Génération de nombres aléatoires suivant une loi binomiale
def loi_binomiale(n):
    # On génère n nombres aléatoires suivant une loi binomiale
    # de paramètres n = 100 et p = 0.5
    x = np.random.binomial(100, 0.5, n)

    return x

# Affichage de l'histogramme
def draw_histogram(f, title):
    fig, axis = plt.subplots(2, 2)

    # N = 10
    axis[0, 0].hist(f(10), bins=50)
    axis[0, 0].set_title("N = 10")
    
    # N = 100
    axis[0, 1].hist(f(100), bins=50)
    axis[0, 1].set_title("N = 100")
    
    # N = 1000
    axis[1, 0].hist(f(1000), bins=50)
    axis[1, 0].set_title("N = 1000")
    
    # N = 10000
    axis[1, 1].hist(f(10000), bins=50)
    axis[1, 1].set_title("N = 10000")

    fig.suptitle(title)
    fig.tight_layout()
    plt.show()

draw_histogram(loi_normale, "Loi normale")
draw_histogram(loi_binomiale, "Loi binomiale")
```
